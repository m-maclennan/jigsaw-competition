```python
# %% Imports & Config

import os, json, math, warnings
import numpy as np
import pandas as pd
from pathlib import Path

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import CalibratedClassifierCV, calibration_curve
from sklearn.metrics import (
    f1_score, brier_score_loss, log_loss, precision_recall_fscore_support
)
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

# --- Reproducibility ---
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# --- Paths ---
DATA_DIR   = Path(".../data/raw")
RESULTS    = Path("results")
RESULTS.mkdir(parents=True, exist_ok=True)

TRAIN_PATH = DATA_DIR / "train.csv"
TEST_PATH  = DATA_DIR / "test.csv"   # not used here
SAMP_PATH  = DATA_DIR / "sample_submission.csv"  # not used here

# --- Columns (adjust if your schema differs) ---
TEXT_COL   = "body"
TARGET_COL = "rule_violation"

# --- TF-IDF params (fast & strong baseline) ---
TFIDF_MAX_FEATURES = 15000
TFIDF_NGRAM_RANGE  = (1, 2)

# --- Plot style ---
plt.rcParams["figure.figsize"] = (7.5, 6)
plt.rcParams["axes.grid"] = True
```

```python
# %% Load Data & Basic Checks

assert TRAIN_PATH.exists(), f"Missing {TRAIN_PATH}"
train_df = pd.read_csv(TRAIN_PATH)

# Schema sanity: fall back if needed
if TEXT_COL not in train_df.columns:
    # Try common alternates seen earlier docs
    for cand in ["comment_text", "text", "body"]:
        if cand in train_df.columns:
            TEXT_COL = cand
            break

if TARGET_COL not in train_df.columns:
    # Try common alternates
    for cand in ["target", "label", "rule_violation"]:
        if cand in train_df.columns:
            TARGET_COL = cand
            break

print(f"Detected TEXT_COL={TEXT_COL}, TARGET_COL={TARGET_COL}")
print("Train shape:", train_df.shape)

# Drop NAs for safety
train_df = train_df[[TEXT_COL, TARGET_COL]].dropna().reset_index(drop=True)
train_df[TARGET_COL] = train_df[TARGET_COL].astype(int)
train_df.head(3)
```

```python
# %% Safety guard: ensuring X_fit, X_cal, X_val exist 


RANDOM_STATE = 42

def _make_splits(df, text_col, target_col):
    X_full = df[text_col].astype(str).values
    y_full = df[target_col].astype(int).values

    # final validation split
    X_tmp, X_val, y_tmp, y_val = train_test_split(
        X_full, y_full, test_size=0.20, random_state=RANDOM_STATE, stratify=y_full
    )
    # split tmp into model-fit and calibration-fit
    X_fit, X_cal, y_fit, y_cal = train_test_split(
        X_tmp, y_tmp, test_size=0.20, random_state=RANDOM_STATE, stratify=y_tmp
    )
    return X_fit, X_cal, X_val, y_fit, y_cal, y_val

need_splits = any(name not in globals() for name in ["X_fit","X_cal","X_val","y_fit","y_cal","y_val"])
if need_splits:
    assert "train_df" in globals(), "train_df not found; run the data-loading cell first."
    assert "TEXT_COL" in globals() and "TARGET_COL" in globals(), "TEXT_COL/TARGET_COL missing; run the schema-detect cell first."
    X_fit, X_cal, X_val, y_fit, y_cal, y_val = _make_splits(train_df, TEXT_COL, TARGET_COL)

print("Split check:",
      f"X_fit={len(X_fit)}  X_cal={len(X_cal)}  X_val={len(X_val)}  |  "
      f"y_fit={len(y_fit)}  y_cal={len(y_cal)}  y_val={len(y_val)}")
```

```python
# %% Vectorise (TF-IDF)
tfidf = TfidfVectorizer(
    max_features=TFIDF_MAX_FEATURES,
    ngram_range=TFIDF_NGRAM_RANGE,
    min_df=2, max_df=0.98
)

X_fit_tfidf = tfidf.fit_transform(X_fit)
X_cal_tfidf = tfidf.transform(X_cal)
X_val_tfidf = tfidf.transform(X_val)

print("TF-IDF shapes:", X_fit_tfidf.shape, X_cal_tfidf.shape, X_val_tfidf.shape)
```

```python
# %% Base Model: Logistic Regression (fast, strong baseline)

base = LogisticRegression(
    solver="lbfgs",
    max_iter=2000,
    class_weight="balanced",
    n_jobs=None,
    random_state=RANDOM_STATE
)
base.fit(X_fit_tfidf, y_fit)

# Uncalibrated probabilities on validation
p_val_uncal = base.predict_proba(X_val_tfidf)[:, 1]
```

```python
# %% Calibrators: Platt (sigmoid) and Isotonic

cal_sigmoid = CalibratedClassifierCV(base, method="sigmoid", cv="prefit")
cal_sigmoid.fit(X_cal_tfidf, y_cal)

cal_isotonic = CalibratedClassifierCV(base, method="isotonic", cv="prefit")
cal_isotonic.fit(X_cal_tfidf, y_cal)

p_val_sigmoid = cal_sigmoid.predict_proba(X_val_tfidf)[:, 1]
p_val_isotonic = cal_isotonic.predict_proba(X_val_tfidf)[:, 1]
```

```python
# %% Metrics helpers

def ece_score(y_true, y_prob, n_bins=15):
    """Expected Calibration Error (ECE) with equal-width bins in [0,1]."""
    y_true = np.asarray(y_true)
    y_prob = np.asarray(y_prob)
    bins = np.linspace(0.0, 1.0, n_bins + 1)
    binids = np.digitize(y_prob, bins) - 1
    ece = 0.0
    for b in range(n_bins):
        mask = binids == b
        if not np.any(mask):
            continue
        conf = y_prob[mask].mean()
        acc  = y_true[mask].mean()
        w    = mask.mean()
        ece += np.abs(acc - conf) * w
    return float(ece)

def f1_at_threshold(y_true, y_prob, thr=0.5):
    y_pred = (y_prob >= thr).astype(int)
    _, _, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
    return float(f1)

def best_threshold_for_f1(y_true, y_prob):
    # Try many thresholds, pick the one with best F1
    ts = np.unique(np.quantile(y_prob, np.linspace(0.01, 0.99, 99)))
    scores = [f1_at_threshold(y_true, y_prob, t) for t in ts]
    i = int(np.argmax(scores))
    return float(ts[i]), float(scores[i])
```

```python
# %% Evaluate Uncalibrated vs Calibrated (on final validation)

def evaluate_variant(name, y_true, y_prob):
    # Default threshold 0.5
    f1_05 = f1_at_threshold(y_true, y_prob, thr=0.5)
    # Best F1 threshold
    best_thr, f1_best = best_threshold_for_f1(y_true, y_prob)
    # Proper scoring rules
    brier = brier_score_loss(y_true, y_prob)
    ll    = log_loss(y_true, np.clip(y_prob, 1e-7, 1 - 1e-7))
    # Calibration quality
    ece = ece_score(y_true, y_prob, n_bins=15)
    return {
        "variant": name,
        "f1@0.5": round(f1_05, 4),
        "best_thr": round(best_thr, 4),
        "f1@best": round(f1_best, 4),
        "brier": round(brier, 6),
        "log_loss": round(ll, 6),
        "ece": round(ece, 6),
    }

metrics = []
metrics.append(evaluate_variant("uncalibrated", y_val, p_val_uncal))
metrics.append(evaluate_variant("sigmoid",      y_val, p_val_sigmoid))
metrics.append(evaluate_variant("isotonic",     y_val, p_val_isotonic))

pd.DataFrame(metrics)
```

```python
# %% Save metrics to CSV (results/calibration_metrics.csv)

metrics_df = pd.DataFrame(metrics).sort_values("brier")
metrics_path = RESULTS / "calibration_metrics.csv"
metrics_df.to_csv(metrics_path, index=False)
print("Saved:", metrics_path.resolve())
metrics_df
```

```python
# %% Reliability Diagrams

def plot_reliability(y_true, y_prob, title, savepath):
    frac_pos, mean_pred = calibration_curve(y_true, y_prob, n_bins=10, strategy="uniform")
    plt.figure()
    plt.plot([0, 1], [0, 1], linestyle="--", label="Perfect")
    plt.plot(mean_pred, frac_pos, marker="o", label="Observed")
    plt.xlabel("Mean predicted probability")
    plt.ylabel("Fraction of positives")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.savefig(savepath, dpi=200)
    plt.close()
    print("Saved:", savepath)

plot_reliability(y_val, p_val_uncal,   "Reliability: Uncalibrated", RESULTS / "reliability_uncalibrated.png")
plot_reliability(y_val, p_val_sigmoid, "Reliability: Platt (sigmoid)", RESULTS / "reliability_sigmoid.png")
plot_reliability(y_val, p_val_isotonic,"Reliability: Isotonic", RESULTS / "reliability_isotonic.png")
```

```python
# %% Show reliability diagrams (side by side)
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

reliability_files = [
    "results/reliability_uncalibrated.png",
    "results/reliability_sigmoid.png",
    "results/reliability_isotonic.png",
]

# Quick existence check
missing = [f for f in reliability_files if not Path(f).exists()]
if missing:
    print("Missing files:", missing)
else:
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    titles = ["Uncalibrated", "Platt (sigmoid)", "Isotonic"]
    for ax, f, t in zip(axes, reliability_files, titles):
        ax.imshow(mpimg.imread(f))
        ax.axis("off")
        ax.set_title(t)
    fig.suptitle("Reliability Diagrams", y=1.05)
    plt.tight_layout()
    plt.show()
```

```python
# %% Probability Histograms (confidence distribution)

def plot_hist(y_prob, title, savepath):
    plt.figure()
    plt.hist(y_prob, bins=25, alpha=0.9)
    plt.xlabel("Predicted probability (positive class)")
    plt.ylabel("Count")
    plt.title(title)
    plt.tight_layout()
    plt.savefig(savepath, dpi=200)
    plt.close()
    print("Saved:", savepath)

plot_hist(p_val_uncal,   "Prob Histogram: Uncalibrated", RESULTS / "hist_uncalibrated.png")
plot_hist(p_val_sigmoid, "Prob Histogram: Platt (sigmoid)", RESULTS / "hist_sigmoid.png")
plot_hist(p_val_isotonic,"Prob Histogram: Isotonic", RESULTS / "hist_isotonic.png")
```

```python
# %% Save validation predictions for documentation

def save_val_preds(name, y_true, y_prob, path):
    df = pd.DataFrame({
        "y_true": y_true,
        "y_prob": y_prob,
    })
    # Also include best-threshold binary predictions
    thr, _ = best_threshold_for_f1(y_true, y_prob)
    df["y_pred_bestF1"] = (df["y_prob"] >= thr).astype(int)
    df.to_csv(path, index=False)
    print("Saved:", path)

save_val_preds("uncalibrated", y_val, p_val_uncal,   RESULTS / "val_predictions_uncalibrated.csv")
save_val_preds("sigmoid",      y_val, p_val_sigmoid, RESULTS / "val_predictions_sigmoid.csv")
save_val_preds("isotonic",     y_val, p_val_isotonic,RESULTS / "val_predictions_isotonic.csv")
```
